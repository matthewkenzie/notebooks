```python
import pandas as pd
import numpy as np

entries = 10000
xrange = (0,1)
yrange = (0,10)

df = pd.DataFrame(columns=['x','y','NN'])

df['x'] = np.random.uniform(*xrange,size=entries)
df['y'] = np.random.uniform(*yrange,size=entries)
df['NN'] = np.random.normal(0.5,0.2,size=entries)

print(df)
```

```python
import matplotlib.pyplot as plt
hn, _, _ = np.histogram2d( df['x'], df['y'], range=(xrange,yrange), bins=(100,100) )
hn[hn==0] = 1
h, xe, ye = np.histogram2d( df['x'], df['y'], range=(xrange,yrange), bins=(100,100), weights=df['NN'] )
fig, ax = plt.subplots(figsize=(6,4))
im = ax.imshow(h.T/hn.T, extent=[xe[0],xe[-1],ye[0],ye[-1]],aspect='auto')
fig.colorbar(im,ax=ax)
fig.tight_layout()
plt.show()
```
